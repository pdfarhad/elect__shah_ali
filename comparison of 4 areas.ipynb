```python
my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k','m']), None, len(df)))
# my_colors = ['g', 'b']*5 
ax = input_file.plot(x=['year','zone'],y=['percentage_change'],kind='bar',color= my_colors)
plt.show()
for p in ax.patches:
    b=p.get_bbox()
    ax.annotate("{:+.2f}".format(b.y1 + b.y0), ((b.x0 + b.x1)/2 - 0.03, b.y1 + 0.02))
```
