```python
def IMPUTE_UNIT(x):
    if x.UNIT < 0:
        return 0
    else:
        return x.UNIT
def IMPUTE_UNIT_WITH_MEAN(x):
    if x.UNIT < 0:
        acc = x.ACCOUNT_NO
        year = x.YEAR
        mean = neg_group.UNIT.get_group((acc,year)).mean()
#         print('{0} {1} {2}'.format(acc, year, mean))
        return mean
    else:
        return x.UNIT
```

```python
df['IMPUTE_UNIT'] = df.apply( IMPUTE_UNIT, axis=1)
```

```python
df.drop('IMPUTE_UNIT', axis=1, inplace=True) # droping the columns with 0s
```

```python
df['IMPUTE_UNIT'] = df.apply(IMPUTE_UNIT_WITH_MEAN,axis=1)
```
